# Circuito RLC

Se tiene un circuito RLC, con unas condiciones iniciales de carga y corriente, se sabe de ante mano que los circuitos RLC son amortiguados, del sistema se tienen las siguientes ecuaciones: <br> 

## Programa para resolver el Problema
Se usará el método de Runge-Kutta 4 


### Librerías Usadas

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

### Constantes del problema

```python
R1 = 350           # Resistencia 1 [Ohm]
R2 = 220           # Resistencia 2 [Ohm]
L1 = 1.2           # Inductancia 1 [H]
L2 = 2             # Inductancia 2 [H]
C1 = 3.5*0.001           # Capacitancia 1 [F]
C2 = 2.8*0.001           # Capacitancia 2 [F]
```

## Algoritmo de runge-kutta 4

```python
f1 = '(-(1/C2)*y1 + (1/C2)*y2 + (1/C2)*y4  + R2*y5 - (1/C1)*y1 + V )/R2'
f2 = 'y3'
f3 = '((1/C2)*y1 - (1/C2)*y2 - (1/C2)*y4)/L2'
f4 = 'y5'
f5 = '(V - (1/C1)*y1 - R1*y5)/L1'
h = 0.001                                   # paso temporal
t = np.arange(0 , 8.001 , 0.001)            # vector de paso temporal
Y1 = np.zeros(len(t))                       # carga electrica 1 
Y2 = np.zeros(len(t))                       # carga electrica 2
Y3 = np.zeros(len(t))                       # corriente electrica 2
Y4 = np.zeros(len(t))                       # carga electrica 3
Y5 = np.zeros(len(t))                       # corriente electrica 3
V = 0                                       # Voltaje [V]
V_1 = np.zeros(len(t))
y1 = 0.001
y2 = 0
y3 = 0 
y4 = 0 
y5 = 0    
for g in range(0,1000):
    y1 = Y1[g]; y2 = Y2[g] ; y3 = Y3[g] ; y4 = Y4[g]; y5 = Y5[g]                                                   
    k1_1 = eval(f1)
    k1_2 = eval(f2)
    k1_3 = eval(f3)
    k1_4 = eval(f4)
    k1_5 = eval(f5)  

    y1 = Y1[g] + (1/2)*k1_1*h ;y2 = Y2[g] + (1/2)*k1_2*h; y3 = Y3[g] + (1/2)*k1_3*h; y4 = Y4[g] + (1/2)*k1_4*h; y5 = Y5[g] + (1/2)*k1_5*h
    k2_1 = eval(f1)
    k2_2 = eval(f2)
    k2_3 = eval(f3)
    k2_4 = eval(f4)
    k2_5 = eval(f5)

    y1 = Y1[g] + (1/2)*k2_1*h ;y2 = Y2[g] + (1/2)*k2_2*h; y3 = Y3[g] + (1/2)*k2_3*h; y4 = Y4[g] + (1/2)*k2_4*h; y5 = Y5[g] + (1/2)*k2_5*h
    k3_1 = eval(f1)
    k3_2 = eval(f2)
    k3_3 = eval(f3)
    k3_4 = eval(f4)
    k3_5 = eval(f5)

    y1 = Y1[g] + k3_1*h ;y2 = Y2[g] + k3_2*h; y3 = Y3[g] + k3_3*h; y4 = Y4[g] + k3_4*h; y5 = Y5[g] + k3_5*h
    k4_1 = eval(f1)
    k4_2 = eval(f2)
    k4_3 = eval(f3)
    k4_4 = eval(f4)
    k4_5 = eval(f5)

    Y1[g + 1] = Y1[g] + (1/6)*(k1_1 + 2*k2_1 + 2*k3_1 + k4_1)*h
    Y2[g + 1] = Y2[g] + (1/6)*(k1_2 + 2*k2_2 + 2*k3_2 + k4_2)*h
    Y3[g + 1] = Y3[g] + (1/6)*(k1_3 + 2*k2_3 + 2*k3_3 + k4_3)*h
    Y4[g + 1] = Y4[g] + (1/6)*(k1_4 + 2*k2_4 + 2*k3_4 + k4_4)*h
    Y5[g + 1] = Y5[g] + (1/6)*(k1_5 + 2*k2_5 + 2*k3_5 + k4_5)*h
for j in range(1001,len(t) - 1):
    V = 110                                                         # Voltaje [V]
    V_1[j] = 110
    y1 = Y1[j]; y2 = Y2[j] ; y3 = Y3[j] ; y4 = Y4[j]; y5 = Y5[j]
    k1_1 = eval(f1)
    k1_2 = eval(f2)
    k1_3 = eval(f3)
    k1_4 = eval(f4)
    k1_5 = eval(f5)
    

    y1 = Y1[j] + (1/2)*k1_1*h ;y2 = Y2[j] + (1/2)*k1_2*h; y3 = Y3[j] + (1/2)*k1_3*h; y4 = Y4[j] + (1/2)*k1_4*h; y5 = Y5[j] + (1/2)*k1_5*h
    k2_1 = eval(f1)
    k2_2 = eval(f2)
    k2_3 = eval(f3)
    k2_4 = eval(f4)
    k2_5 = eval(f5)


    y1 = Y1[j] + (1/2)*k2_1*h ;y2 = Y2[j] + (1/2)*k2_2*h; y3 = Y3[j] + (1/2)*k2_3*h; y4 = Y4[j] + (1/2)*k2_4*h; y5 = Y5[j] + (1/2)*k2_5*h
    k3_1 = eval(f1)
    k3_2 = eval(f2)
    k3_3 = eval(f3)
    k3_4 = eval(f4)
    k3_5 = eval(f5)
    

    y1 = Y1[j] + k3_1*h ;y2 = Y2[j] + k3_2*h; y3 = Y3[j] + k3_3*h; y4 = Y4[j] + k3_4*h; y5 = Y5[j] + k3_5*h
    k4_1 = eval(f1)
    k4_2 = eval(f2)
    k4_3 = eval(f3)
    k4_4 = eval(f4)
    k4_5 = eval(f5)
    
    Y1[j + 1] = Y1[j] + (1/6)*(k1_1 + 2*k2_1 + 2*k3_1 + k4_1)*h
    Y2[j + 1] = Y2[j] + (1/6)*(k1_2 + 2*k2_2 + 2*k3_2 + k4_2)*h
    Y3[j + 1] = Y3[j] + (1/6)*(k1_3 + 2*k2_3 + 2*k3_3 + k4_3)*h
    Y4[j + 1] = Y4[j] + (1/6)*(k1_4 + 2*k2_4 + 2*k3_4 + k4_4)*h
    Y5[j + 1] = Y5[j] + (1/6)*(k1_5 + 2*k2_5 + 2*k3_5 + k4_5)*h
# CALCULO DE LA CORRIENTE 1  
V_1[len(V_1)-1] = 110
Y6 = (-(1/C2)*Y1 + (1/C2)*Y2 + (1/C2)*Y4  + R2*Y5 - (1/C1)*Y1 + V_1 )/R2   # CORRIENTE 1
```

```python
plt.figure(figsize=(5*1.618,5))
plt.plot(t,Y6, color='k', label='$I_1(t)$', linewidth=2)
plt.legend()
plt.xlabel('$Tiempo [s]$',fontsize=12)
plt.ylabel('$Corriente 1 [A]$',fontsize=12)
plt.grid()
plt.figure(figsize=(5*1.618,5))
plt.plot(t,Y3, color='r', label='$I_2(t)$', linewidth=2)
plt.legend()
plt.xlabel('$Tiempo [s]$',fontsize=12)
plt.ylabel('$Corriente 2 [A]$',fontsize=12)
plt.grid()
plt.figure(figsize=(5*1.618,5))
plt.plot(t,Y5, color='b', label='$I_3(t)$', linewidth=2)
plt.legend()
plt.xlabel('$Tiempo [s]$',fontsize=12)
plt.ylabel('$Corriente 3 [A]$',fontsize=12)
plt.grid()
```
